```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('../goal_imgs/goals.pkl', 'rb') as f:
    goals = pickle.load(f)
    # x range of data ~0.9368 - 1.2285
```

```python
def show_img(idx):
    img = goals[idx][0]
    context = goals[idx][1]
    gt = goals[idx][2]
    plt.imshow(img[0][0], cmap="gray")
    return (img, context, gt)

def search_idx_closest_to_gt(value, axis=0):
    old_diff = 1000
    for ii in range(len(goals)):
        diff = abs(goals[ii][2][axis].item() - value)
        if diff < old_diff:
            idx = ii
            old_diff = diff
    return idx
```

```python
idx = search_idx_closest_to_gt(100)
goals[idx][2]
data = show_img(idx)
```

```python
with open('../goal_imgs/goal_medium.pkl', 'wb') as handle:
    pickle.dump(data, handle)
```
